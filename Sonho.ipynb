```python
import csv
import requests
from bs4 import BeautifulSoup

url = 'https://www.factor.gg/teams'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find(class_='w-full stats-table')
rows = table.find_all('tr')

# pega a primeira linha da tabela
header_row = rows[0]

# obtÃ©m todas as cÃ©lulas da primeira linha e imprime o texto de cada cÃ©lula
cells = header_row.find_all('th')
row_values = []

for cell in cells:
    # obtÃ©m o texto da cÃ©lula
    cell_text = cell.text.strip()
    
    # remove tudo apÃ³s o segundo espaÃ§o
    cell_text = ' '.join(cell_text.split(' ')[:2])
    
    # remove a primeira ocorrÃªncia da palavra "POWR"
    cell_text = cell_text.replace('POWR ', '', 1)
    
    # remove todas as ocorrÃªncias das palavras escolhidas
    cell_text = cell_text.replace('Index', '').replace('Average', '')
    cell_text = cell_text.replace('First Blood', '').replace('First Dragon', '').replace('First Tower', '')
    
    row_values.append(cell_text)

# escreve os valores da linha em um arquivo CSV
with open('output.csv', mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file, delimiter=';')
    writer.writerow(row_values)

df= pd.read_csv('output.csv', sep=";")
display(df)
```

```python
import csv

# Abrir os arquivos CSV como listas
with open('teams.csv', 'r', newline='') as f_teams, open('output.csv', 'r', newline='') as f_output:
    reader_teams = csv.reader(f_teams, delimiter=';')
    reader_output = csv.reader(f_output, delimiter=';')
    data_teams = [row for row in reader_teams]
    data_output = [row for row in reader_output]

# Salvar a primeira linha do arquivo output.csv em uma variÃ¡vel
output_first_row = data_output[0]

# Adicionar a primeira linha do arquivo output.csv como a primeira linha do arquivo teams.csv
data_teams.insert(0, output_first_row)

# Escrever a lista combinada de volta em teams.csv
with open('teams.csv', 'w', newline='') as f_teams:
    writer_teams = csv.writer(f_teams, delimiter=';')
    writer_teams.writerows(data_teams)
    
df= pd.read_csv('teams.csv', sep=";")
display(df)
```
