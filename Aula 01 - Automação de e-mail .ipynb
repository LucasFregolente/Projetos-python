```python
#--------------------------------------------------------------------------------------------------------------

# pyoutogui é a biblioteca de controle de perifericos (teclado e mouse)
import pyautogui as py

# time é a biblioteca que adiciona comandos de delay - (com a exeção de py.PAUSE) 
import time

#--------------------------------------------------------------------------------------------------------------
# comandos basicos a serem gravados
#--------------------------------------------------------------------------------------------------------------

# import biblioteca as apelido - abrevia a chamada da biblioteca

# biblioteca pyautogui
# py.click() - clica na cordenada indicada
# py.write() - escreve o que ta escrito
# py.press() - pressiona uma unica tecla
# py.hotkey() - pressiona uma combinação de teclas (ctrl + v)

#--------------------------------------------------------------------------------------------------------------

# definindo delay para os comando da bibliotecas pyautogui

py.PAUSE = 1

#--------------------------------------------------------------------------------------------------------------
# PASSO 1: Entrar no drive
# OBS: https://drive.google.com/drive/my-drive
#--------------------------------------------------------------------------------------------------------------

# nova guia
py.hotkey("ctrl","t")

# link do drive
py.write("https://drive.google.com/drive/my-drive")
py.press("enter")

# delay para carregar a guia do navegador
time.sleep(5)

# duplo click na pasta do banco de dados
py.doubleClick(x=368, y=526)

# delay para carregar a guia do navegador
time.sleep(5)

#--------------------------------------------------------------------------------------------------------------
# PASSO 2: Baixar bando de dados
#--------------------------------------------------------------------------------------------------------------
# clicando no arquivo
py.click(x=385, y=455)

# clicando nos 3 pontos
py.click(x=1717, y=162)

# clicando nos download
py.click(x=1528, y=503)

# delay para baixar o arquivo
time.sleep(5)

#--------------------------------------------------------------------------------------------------------------
```

```python
#--------------------------------------------------------------------------------------------------------------
# PASSO 3: Calcular indicadores 
#--------------------------------------------------------------------------------------------------------------

# pandas é a biblioteca para manipulação e análise de dados
import pandas as pd 

#--------------------------------------------------------------------------------------------------------------

# carregando a tabela
tabela= pd.read_csv(r"C:\Users\A\Downloads\Compras.csv", sep=";")

# mostra a tabela de um forma organizada (COMANDO DO JUPYTER "display")
display(tabela)

# calculo dentro da tabela 

# colocando a soma total do  da coluna "......" na variavel
total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto/quantidade

# mostrando as variaveis
print(total_gasto)
print(quantidade)
print(preco_medio)

#--------------------------------------------------------------------------------------------------------------
```

```python
#--------------------------------------------------------------------------------------------------------------
# PASSO 4: Mandar no e-mail
#--------------------------------------------------------------------------------------------------------------

import pyautogui as py
import pyperclip as pc

# criar uma nova guia
py.hotkey("ctrl","t")

# entrar no e-mail: https://mail.google.com/mail/u/0/#inbox
py.write("https://mail.google.com/mail/u/0/#inbox")
py.press("enter")

# delay para carregar a guia do navegador
time.sleep(5)

# clicar no botão escrever
py.click(x=151, y=171)
time.sleep(1)

# colocar destinatario 
py.write("lucasfregolente02@gmail.com")

# tab para finalizar o primeiro e unico destinatario
py.press("tab")

# para passar para o assunto
py.press("tab")

# colacar o assunto

# OBSERVAÇÃO: O comando copy da biblioteca pyperclip, copia o texto
# de forma que podemos usar a acentuação que nao vem por padrão. 

pc.copy("PROJETO AUTOMAÇÃO COM PYTHON")
py.hotkey("ctrl","v")

# para passar para o corpo do texto
py.press("tab")

# colocar o corpo do texto com os dados

# OBS: salvando o texto em uma variavel para facilitar o desenvolvimento
# f indica que é um texto com variaveis
# : indica formatação na variavel
# , indica que o separador sera a virgula
# .2f indica que a variavel terá 2 casas decimais

texto = f"""
Prezado eu,
Segue o resumo do relatório  de compras:

Gasto Total: R$ {total_gasto:,.2f}
Quantidade: {quantidade:,}
Preço Médio: R$ {preco_medio:,.2f}

Dados obtidos pelo processo automatizado em python

Att.,
Lucas Fregolente"""

#copiando  e colando o "texto" no e-mail
pc.copy(texto)
py.hotkey("ctrl","v")


# enviar
py.hotkey("ctrl","enter")
```
